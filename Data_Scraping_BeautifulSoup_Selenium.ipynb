```python
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
```

```python
jobs = soup.find_all("li", class_ = "clearfix job-bx wht-shd-bx")
```

```python
len(jobs)
```

```python
jobs_dict = {"published_date":[],"company_name":[],"skills":[],"job_description":[]}
for job in tqdm(jobs):
    
    published_date = job.find("span", class_ = "sim-posted").span.text
    company_name = job.find("h3" , class_ = "joblist-comp-name").text.replace(" ", "")
    skills = job.find("span", class_ = "srp-skills").text.replace(" ","")
    job_description = job.header.h2.a["href"]
    jobs_dict['published_date'].append(published_date)
    jobs_dict['company_name'].append(company_name)
    jobs_dict['skills'].append(skills)
    jobs_dict['job_description'].append(job_description)
```
