```python

def extract_info():
    quotes_dict = {"quote":[],"author":[],'author_url':[],'tags':[]}

    for single_quote in quote_divs:
#         print(quote_divs)
        single_quote_txt = single_quote.span.string
        author = single_quote.span.next_sibling.next_sibling.small.string
        tags = single_quote.div
        url = single_quote.span.next_sibling.next_sibling.a['href']

        tags_list = []

        quotes_dict["quote"].append(single_quote_txt)
        quotes_dict['author'].append(author)
        quotes_dict['author_url'].append(url)
        for anchor in tags.find_all('a'):
            tags_list.append(anchor.string)
        quotes_dict['tags'].append(tags_list)
    
    return pd.DataFrame(quotes_dict)
```
